Se cagra la paqueteria necesaria para la regreseion lineal multiple.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

Se carga el dataset con pandas

```python
data = pd.read_csv("real_estate_price_size_year.csv")
data
```

Podemos ver los valores estadisticos relevantes

```python
data.describe()
```

Crearemos la regresion.
Primero declaramos las variables dependientes e independientes

```python
y = data['price']
x1 = data[['size','year']]
```

Iniciamos la regresion

```python
x = sm.add_constant(x1) # Agrega una columna de unos a la matriz constituida por x1
results = sm.OLS(y,x).fit() #Utiliza la paqueteria de statsmodels para generear la regresion lineal utilizando el metodo de Minimos Cuadrados Oridnarios, luego ajusta una recta con el resultado.
results.summary() # Nos da los resultados de la regresion lineal.
```

Ahora, podemos analizar los resultados obtenidos con la regresión linal multiple. Como podemos ver en el cuadro con los resultados de la regresión. Nuestras variables 'size' y 'year' tienen un valor P igual a 0 por lo que podemos concluir que son significativas para la regresión. Además, tenemos un R^2 relativamente alto lo cual nos indica que nuestra regresión explica una buena variabilidad total. Por último, tenemos un valor alto en nuestra estadística F lo cual, por consiguiente, nos da practicamente 0 en el valor de F ajustado, esto nos dice que nuestro modelo es significativo.
